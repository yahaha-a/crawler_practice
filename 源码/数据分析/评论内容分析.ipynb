```python
import jieba
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# 如果浏览器不显示图片，就需要加上这句话
%matplotlib inline  

# 让图片中可以显示中文
plt.rcParams['font.sans-serif'] = 'SimHei'   
# 让图片中可以显示负号
plt.rcParams['axes.unicode_minus'] = False  

# 支持svg矢量图
%config Inlinebackend.figure_format = 'svg'  
```

```python
with open('./评论内容/高考_评论内容.txt', "r", encoding='utf-8') as f:
    text = f.read() 

words = jieba.lcut(text.strip())
counts = {}
stopwords = [line.strip() for line in open('./停用词.txt', 'r', encoding='utf-8').readlines()]
for word in words:
    if len(word) == 1:
        continue
    elif word not in stopwords:
        counts[word] = counts.get(word, 0) + 1
items = list(counts.items())
items.sort(key=lambda x: x[1], reverse=True)  # 排序
f = open('./评论内容/高考_词频.txt', "w", encoding='utf-8')
for i in range(20):
    word, count = items[i]
    f.writelines("{}\t{}\n".format(word, count))
f.close()
```

```python
with open("./评论内容/高考_词频.txt", 'r', encoding='utf-8') as f:
    text_高考 = f.read()
with open("./评论内容/考研_词频.txt", 'r', encoding='utf-8') as f:
    text_考研 = f.read()
with open("./评论内容/考公_词频.txt", 'r', encoding='utf-8') as f:
    text_考公 = f.read()
with open("./评论内容/就业_词频.txt", 'r', encoding='utf-8') as f:
    text_就业 = f.read()

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

wcloud = WordCloud(
    font_path='./NotoSansSC-Light.otf',
    background_color="white",
    width=1000,
    max_words=500,
    height=860,
    margin=2
).generate(text_高考)

axs[0, 0].axis('off')
axs[0, 0].imshow(wcloud)
axs[0, 0].set_title('高考')

wcloud = WordCloud(
    font_path='./NotoSansSC-Light.otf',
    background_color="white",
    width=1000,
    max_words=500,
    height=860,
    margin=2
).generate(text_考研)

axs[0, 1].axis('off')
axs[0, 1].imshow(wcloud)
axs[0, 1].set_title('考研')

wcloud = WordCloud(
    font_path='./NotoSansSC-Light.otf',
    background_color="white",
    width=1000,
    max_words=500,
    height=860,
    margin=2
).generate(text_考公)

axs[1, 0].axis('off')
axs[1, 0].imshow(wcloud)
axs[1, 0].set_title('考公')

wcloud = WordCloud(
    font_path='./NotoSansSC-Light.otf',
    background_color="white",
    width=1000,
    max_words=500,
    height=860,
    margin=2
).generate(text_就业)

axs[1, 1].axis('off')
axs[1, 1].imshow(wcloud)
axs[1, 1].set_title('就业')

plt.tight_layout()
plt.show()
```
